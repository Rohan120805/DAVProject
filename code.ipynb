```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("insurance.csv")
```

```python
#df['region'].unique()
df.isnull().sum()
```

```python
df['age'].unique()
```

```python
plt.figure(figsize=(7, 7))
df['bmi'].plot(kind='hist', bins=30, edgecolor='black')
plt.xlabel('BMI')
plt.title('Distribution of BMI')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['age'].plot(kind='hist', bins=30, edgecolor='black')
plt.xlabel('AGE')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(7, 7))
df['children'].plot(kind='hist', bins=30, edgecolor='black')
plt.xlabel('Number of Children')
plt.title('Frequency of BMI')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x=df["age"],y=df["charges"])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['smoker'], y=df['charges'])
plt.title('Charges comparision of  smokers and non-smokers')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['region'], y=df['smoker'])
plt.title('smokers and non-smokers in each region')
plt.show()
```
