```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("insurance.csv")
```

```python
#df['region'].unique()
df.isnull().sum()
```

```python
df['region'].unique()
```

```python
plt.figure(figsize=(7, 7))
df['bmi'].plot(kind='hist', bins=30, edgecolor='black')
plt.xlabel('BMI')
plt.title('Distribution of BMI')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['bmi'], df['charges'], c='blue', alpha=0.5)
plt.title('BMI vs Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()
```

```python
plt.hist(df['age'], bins=15, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(7, 7))
df['children'].plot(kind='hist', bins=30, edgecolor='black')
plt.xlabel('Number of Children')
plt.title('Frequency of children')
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['age'], df['charges'], c='red', alpha=0.5)
plt.title('Age vs Insurance Purchase')
plt.xlabel('Age')
plt.ylabel('Charge of Insurance Premium based on Age')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x=df['smoker'], y=df['charges'])
plt.title('Charges comparision of  smokers and non-smokers')
plt.show()
```

```python
smoker_counts = df.groupby(['region', 'smoker']).size().unstack(fill_value=0)

smoker_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Number of Smokers and Non-Smokers in Each Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Smoker')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.boxplot(x=df['sex'], y=df['charges'])
plt.title('Charges comparision of Male and Female')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='region', y='charges', data=df, estimator=lambda x: sum(x) / len(x))
plt.title('Average Charges by Region')
plt.xlabel('Region')
plt.ylabel('Average Charges')
plt.show()
```
